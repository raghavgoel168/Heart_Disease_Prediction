```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn. linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# Naive Bayes
from sklearn.naive_bayes import GaussianNB

#ANN and RNN
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

```

```python
df=pd.read_csv("C:/Users/ragha/OneDrive/Desktop/Ethics/Heart_Disease_Prediction.csv")
```

```python
df['Heart Disease']=df['Heart Disease'].map({ 'Presence' : 1, 'Absence' : 0})
df.head()
```

```python
x=df.drop(['Heart Disease' ] ,axis=1) 
y=df[ 'Heart Disease' ]
```

```python
pca=PCA(13)
x_pca=pca.fit_transform(x)
x_pca.shape
```

```python
xtrain_pca, xtest_pca, ytrain,ytest=train_test_split(x_pca,y,test_size=0.4,random_state=11)
```

```python
def report(actual,pred) :
    acc_score=accuracy_score(actual, pred)
    con_mat=confusion_matrix(actual, pred )
    class_rep=classification_report(actual,pred) 
    print("Accuracy of model is", round((acc_score*100),2),'%')
    print( '\n',con_mat) 
    print( ' \n ' , class_rep)
```

```python
log=LogisticRegression(max_iter=1000) 
log.fit(xtrain_pca,ytrain) 
train_log=log.predict (xtrain_pca) 
test_log=log. predict (xtest_pca )
# report(ytrain,train_log) 
report (ytest, test_log)
log.score(xtest_pca, ytest)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(ytest, test_log)
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(ytest, test_log))
print(f'Root Mean Square Error (RMSE): {rmse}')
```

```python
dt=DecisionTreeClassifier(max_depth=7,min_samples_split=17,min_samples_leaf=10)
dt.fit (xtrain_pca, ytrain) 
pred_dtree_train=dt.predict(xtrain_pca) 
pred_dtree_test=dt.predict (xtest_pca ) 
# report(ytrain,pred_dtree_train) 
report (ytest, pred_dtree_test )
dt.score(xtest_pca, ytest)
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(ytest, pred_dtree_test)
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(ytest, pred_dtree_test))
print(f'Root Mean Square Error (RMSE): {rmse}')
```

```python
rf=RandomForestClassifier( n_estimators=15 , max_depth=4, min_samples_split=15,random_state=786)
rf.fit(xtrain_pca,ytrain) 
train_rf=rf.predict(xtrain_pca)  
test_rf=rf.predict (xtest_pca) 
# report(ytrain,train_rf) 
report (ytest, test_rf)
rf.score(xtest_pca, ytest)
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(ytest, test_rf)
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(ytest, test_rf))
print(f'Root Mean Square Error (RMSE): {rmse}')
```

```python
svm_ = SVC(kernel='linear', random_state=0)  
svm_.fit(xtrain_pca, ytrain)
train_svm_=svm_.predict(xtrain_pca)  
test_svm_=svm_.predict (xtest_pca) 
# report(ytrain,train_svm_) 
report (ytest, test_svm_)
svm_.score(xtest_pca, ytest)
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(ytest, test_svm_)
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(ytest, test_svm_))
print(f'Root Mean Square Error (RMSE): {rmse}')
```

```python
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(xtrain_pca,ytrain)
train_knn=knn.predict(xtrain_pca)  
test_knn=knn.predict (xtest_pca) 
# report(ytrain,train_knn) 
report (ytest, test_knn)
knn.score(xtest_pca,ytest)
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(ytest, test_knn)
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(ytest, test_knn))
print(f'Root Mean Square Error (RMSE): {rmse}')
```

```python
# Define the Artificial Neural Network (ANN)
model_ann = Sequential([
    Dense(64, activation='relu', input_dim=xtrain_pca.shape[1]),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model_ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model_ann.fit(xtrain_pca, ytrain, epochs=50, batch_size=16, validation_data=(xtest_pca, ytest), verbose=2)

# Evaluate the model
y_pred_ann_prob = model_ann.predict(xtest_pca)
y_pred_ann = (y_pred_ann_prob > 0.5).astype(int)
# y_pred_ann = model_ann.predict_classes(xtest_pca)

report (ytest, y_pred_ann)
loss_ann, accuracy_ann = model_ann.evaluate(xtest_pca, ytest)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(ytest, y_pred_ann)
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(ytest, y_pred_ann))
print(f'Root Mean Square Error (RMSE): {rmse}')
```

```python
xtrain_rnn = xtrain_pca.reshape((xtrain_pca.shape[0], 1, xtrain_pca.shape[1]))
xtest_rnn = xtest_pca.reshape((xtest_pca.shape[0], 1, xtest_pca.shape[1]))

# Define the Recurrent Neural Network (RNN)
model_rnn = Sequential([
    tf.keras.layers.SimpleRNN(32, activation='relu', input_shape=(xtrain_rnn.shape[1], xtrain_rnn.shape[2])),
    Dense(1, activation='sigmoid')
])

# Compile the model
model_rnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model_rnn.fit(xtrain_rnn, ytrain, epochs=50, batch_size=16, validation_data=(xtest_rnn, ytest), verbose=2)

# Evaluate the model
y_pred_rnn_prob = model_rnn.predict(xtest_rnn)
y_pred_rnn = (y_pred_rnn_prob > 0.5).astype(int)

report (ytest, y_pred_rnn)
accuracy_rnn = model_rnn.evaluate(xtest_rnn, ytest)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(ytest, y_pred_rnn)
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(ytest, y_pred_rnn))
print(f'Root Mean Square Error (RMSE): {rmse}')
```
